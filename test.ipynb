```python
csv_list = [
    'outputs/2025-02-24/22-35-48/metrics.csv',
    'outputs/2025-02-24/22-37-05/metrics.csv',
    'outputs/2025-02-24/22-37-24/metrics.csv',
    # 'outputs/2025-02-24/22-41-47/metrics.csv',
    # 'outputs/2025-02-24/22-49-46/metrics.csv',
]

import pandas as pd
# Read all the csv files into a list of dataframes
dfs = []
for csv in csv_list:
    dfs.append(pd.read_csv(csv))
```

```python
# plot line plots of the 'eval_win_rate_strong' column of all the dfs in a single line plot. label the lines with the index of the dataframe in the list
import matplotlib.pyplot as plt

for i, df in enumerate(dfs):
    plt.plot(df['eval_win_rate_strong'], label=f'DataFrame {i}')

plt.xlabel('Index')
plt.ylabel('Eval Win Rate Strong')
plt.title('Eval Win Rate Strong for All DataFrames')
plt.legend()
plt.show()
```

```python
import tueplots
from tueplots import bundles
from tueplots.constants.color import rgb

colors = [ rgb.tue_lightblue, rgb.tue_darkgreen, rgb.tue_violet, rgb.tue_orange ]
labels = ['TD3 base', '+ PER', '+ Pink Noise']

with plt.rc_context({**bundles.icml2022(column='half', nrows=1, ncols=1, usetex=False), **tueplots.axes.lines()}):

    fig, ax = plt.subplots()

    for i, df in enumerate(dfs):
        ax.plot(
            df['train_episode'],
            df['eval_win_rate_strong'],
            label=labels[i],
            color=colors[i],
            marker='.'
        )

    ax.set_xlabel("Train episodes")
    ax.set_ylabel("Win Rate (%)")
    # ax.set_title("Percentage of people with hesitation over time")
    ax.set_ylim([0, 1.08])

    # Add spines
    ax.spines["top"].set(visible = False)
    ax.spines["right"].set(visible = False)
    ax.grid(True, color = "lightgrey", ls = ":")

    # # Set ticks
    # ax.set_xticks(range(0, 10000, 500))

    plt.legend()
    plt.savefig("outputs/ablation.pdf")
    plt.show()
```

```python
# plot line plots of the 'eval_win_rate_strong' column of all the dfs in a single line plot. label the lines with the index of the dataframe in the list
import matplotlib.pyplot as plt

csv_list = [
    'outputs/2025-02-24/22-49-46/metrics.csv',
    'outputs/2025-02-24/22-41-47/metrics.csv',
    
]

dfs_2 = []
for csv in csv_list:
    dfs_2.append(pd.read_csv(csv))

for i, df in enumerate(dfs_2):
    plt.plot(df['eval_win_rate_strong'], label=f'DataFrame {i}')

plt.xlabel('Index')
plt.ylabel('Eval Win Rate Strong')
plt.title('Eval Win Rate Strong for All DataFrames')
plt.legend()
plt.show()
```

```python
# Create a nice plot for dfs_2 using tueplots
labels_2 = ['TD3 base', '+ Curriculum']

colors_1 = [ rgb.tue_lightblue, rgb.tue_orange]

with plt.rc_context({**bundles.icml2022(column='half', nrows=1, ncols=1, usetex=False), **tueplots.axes.lines()}):

    fig, ax = plt.subplots()

    for i, df in enumerate(dfs_2):
        # Get only the first half of the data
        half_len = int(len(df) * 0.75)
        df_half = df.iloc[:half_len]
        
        ax.plot(
            df_half['train_episode'],
            df_half['eval_win_rate_strong'],
            label=labels_2[i],
            color=colors_1[i],  # Using different colors from the first plot
            marker='.'
        )

    ax.set_xlabel("Train episodes")
    ax.set_ylabel("Win Rate (%)")
    ax.set_ylim([0, 1.08])

    # Add spines
    ax.spines["top"].set(visible = False)
    ax.spines["right"].set(visible = False)
    ax.grid(True, color = "lightgrey", ls = ":")

    plt.legend()
    plt.savefig("outputs/extended_training_first_half.pdf")
    plt.show()
```
